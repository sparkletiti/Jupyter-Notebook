Write a Pandas program to get those movies whose revenue more than 2 million and spent less than 1 million. Let the framework only display the following columns: 'title', 'release_date', 'budget', 'revenue', 'runtime'

The revenue and budget columns need to be changed to numeric values and the error in the revenue column needs to be removed.

```python
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['budget'] = pd.to_numeric(df['budget'])
df[((df['revenue']>2000000) & (df['budget']<=1000000))][['title', 'release_date', 'budget', 'revenue', 'runtime']]
```
